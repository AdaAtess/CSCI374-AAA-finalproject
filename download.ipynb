# Download Images from Instagram Python Script

```python
# Use instaloader to download images from Instagram
from instaloader import Instaloader, Profile
# Create an instance of Instaloader class
L = Instaloader()
```

```python
# Identify the profile to scrape from
PROFILE = "yikyakyeo"
# Get the profile context from profile
profile = Profile.from_username(L.context, PROFILE)
```

```python
# Get the posts from the profile sorted the number of likes
posts_sorted_by_likes = sorted(profile.get_posts(), key=lambda p: p.likes, reverse=True)
```

```python
# Download all the posts and store them into a folder named after the profile
for post in posts_sorted_by_likes:
    L.download_post(post, PROFILE)
```
